Step 1

```python
import pandas as pd
```

Step 2 and 3

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|') 
```

Step 4

```python
def mean_age_based_on_occupation():
    mean_age = users[['occupation', 'age']].groupby('occupation').mean()
    print(mean_age)
mean_age_based_on_occupation()
```

Step 5 - Male ratio per occupation and sort it from the most to the least

```python
#getting the data for gender and occupation
dfgo = users[['gender', 'occupation']]
male_cases = dfgo.loc[dfgo['gender'] == 'M']

#calculating the ratio
group_by_occupation = dfgo.groupby('occupation').count()
male_count = male_cases.groupby('occupation').count()
male_ratio = male_count / group_by_occupation
male_ratio.sort_values(ascending=False, by='gender')
```

Step 6. For each occupation, calculate the minimum and maximum ages

```python
dfoa = users[['occupation', 'age']]

min_age = dfoa.groupby('occupation').min()

max_age = dfoa.groupby('occupation').max()

age_table = pd.concat([max_age, min_age], axis=1)

print(age_table)
```

Step 7. For each combination of occupation and sex, calculate the mean age

```python
dfc = users[['occupation', 'gender', 'age']]

mean_age = dfc.groupby(['occupation', 'gender']).mean('age')

print(mean_age)
```

Step 8. For each occupation present the percentage of women and men

```python
gender_ratio = dfc.groupby('occupation')['gender'].value_counts(normalize=True)*100

print(gender_ratio)
```

```python
data.index = pd.to_datetime(pd.concat([data['Year'], data['Month'], data['Day']], axis=1))
data.drop(['Year','Month','Day'], axis=1, inplace=True) # dropped first 3 columns to avoid redundancy

data
```

Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean
windspeeds and standard deviations of the windspeeds at each location over all the
days
A different set of numbers for each location.

```python
loc_stats = pd.concat([data.min(), data.max(), data.mean(), data.std()], axis=1) 
loc_stats.rename(columns={0:'Minimum', 1:'Maximum', 2:'Mean', 3:'Std_dev'}, inplace=True) 

loc_stats
```

Step 10. Create a DataFrame called day_stats and calculate the min, max and mean
windspeed and standard deviations of the windspeeds across all the locations at each
day.
A different set of numbers for each day.

```python
day_stats = pd.concat([data.min(axis=1), data.max(axis=1), data.mean(axis=1), data.std(axis=1)], axis=1)
day_stats.rename(columns={0:'Minimum',1:'Maximum',2:'Mean',3:'Std_dev'}, inplace=True)

day_stats
```

Step 11. Find the average windspeed in January for each location.
Treat January 1961 and January 1962 both as January.

```python
average_wspeed = data[data.index.month == 1].mean()
average_wspeed
```

Step 13. Downsample the record to a monthly frequency for each location.

```python
monthly_freq = data.resample('M').ffill()

monthly_freq
```

Step 14. Downsample the record to a weekly frequency for each location

```python
weekly_freq = data.resample('W').ffill()

weekly_freq
```

Step 15. Calculate the min, max and mean windspeeds and standard deviations of the
windspeeds across all locations for each week (assume that the first week starts on
January 2 1961) for the first 52 weeks.

```python
weekly_data = weekly_freq[:52]
weekly_data.shape[0]

weekly_data = pd.concat([weekly_data.min(), weekly_data.max(), weekly_data.mean(), weekly_data.std()], axis=1)
weekly_data.rename(columns={0:'Minimum',1:'Maximum',2:'Mean',3:'Std_dev'},inplace=True)

weekly_data
```

Question 6
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

df_div = pd.read_csv('us-marriages-divorces-1867-2014.csv')

df_div
```

```python
plot = plt.figure()

plt.rcParams["figure.figsize"] = [12, 12]

df_div.plot( 'Year', ['Marriages_per_1000','Divorces_per_1000'], label=['Marriages', 'Divorces'])

# Set the Title
plt.title('Marriages and Divorces in the U.S. between 1867 and 2014', color = 'm', fontsize = 20)

# Set x & y axis labels
plt.xlabel('Year', color = 'k', fontsize = 18)
plt.ylabel('Capita', color = 'k', fontsize = 18)

plt.show()
```
